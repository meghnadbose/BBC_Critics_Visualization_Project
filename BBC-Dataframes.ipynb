### Okay, now to explore: 

Right now there's only one Field with the country--the critics country. Let's group by that.

```python
df.groupby('critic_country')['critic_name'].nunique().sort_values(ascending=False)
```

```python
#Narrowing down my data into a smaller frame
```

```python
df1 = df.groupby('critic_country')['movie_name'].value_counts().reset_index(name='count')
df1.tail(50)
```

The cell above has the output I want,  but I need it to look nice and be combined in a single column.  This is how we build the **"article:"** field of our geojson doc. You need to combine columns of data into readable text.

```python
#Moving the values from the two right columns into a new column
#One that human readers can understand
df1["string"] = df1["movie_name"] + ": " + df1["count"].astype(str) + np.where(df1["count"]>1, ' votes', ' vote')
df1.head(40)
```

This is nice but I need to I have only **one row per country**. I do another group, and combine everything together. And when I do that I throw some HTML into there.

```python
#this is just two different ways to do this:
#one using % as the wild card
#and one using .format and {0} as the wild card

#output = df1.groupby('critic_country')['string'].apply(lambda x: "<div id='movie'><h1><b>Top Movies</b></h1><P>%s</P></div>" % '</p><p> '.join(x)).reset_index(name='properties.article')
output = df1.groupby('critic_country')['string'].apply(lambda x: "<div class='movie_list'><h1><b>Top Movies</b></h1><P>{0}</P></div>".format('</p><p> '.join(x))).reset_index(name='properties.article')
output
```

```python
output.iloc[3]['properties.article']
```

Now we need our **headline**:  I'm going to have it be the number of critics in the country. I make a mini dataframe That counts the number of critics, and I merge it with my output dataframe.

```python
crits = df.groupby('critic_country')['critic_name'].nunique().reset_index(name='properties.headline')
crits
```

```python
output = output.merge(crits, how='left', on='critic_country')
```

```python
output
```

```python
#Turn that number and into something readable.
#Though here I don't pay any attention to the English language
#output['properties.headline'] = output['properties.headline'].map(str) + " critic"
output['properties.headline'] = output['properties.headline'].astype(str) + np.where(output["properties.headline"]>1, ' critics', ' critic')
```

```python
output
```

```python
#Add some color (this not the way to do this!! but I'm just showing you the column.
#the color #35476E is a hexadecimal code for colors 35 is the amount of Red, 47 Green, 6E Blue 
# #FFFFFF is white and #000000 is black
output['properties.color'] = "#35476E"
output
```

Here's how you extract the directors from your data frame, to look for more info.

```python
#df.sort_values(by=('director'))
d_list = list(df['movie_dir_split'])
d_list
```

```python
urls = []
for name in d_list:
    dir_each_name = name.split(" ")
    name_segments = name.replace(" ","%20")
    search_url = "https://www.imdb.com/find/?q=" + name_segments
    dic_director = {'director_name': name, 'director_url': search_url}
    urls.append(dic_director)
    print(dic_director)
```

```python
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.182 Safari/537.36"}
```

```python
for url in urls:
    print(url['director_url'])
```

```python
all_dir_urls = []
for url in urls:
    search_url_for_requests = url['director_url']
    response = requests.get(search_url_for_requests, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')

    dir_url_element = soup.find('a', class_="ipc-metadata-list-summary-item__t")
    
    if dir_url_element:
        dir_url_segment = dir_url_element.get('href')
        if dir_url_segment:
            dir_url = "https://www.imdb.com/" + dir_url_segment
            dir_url_dict = {'director_name': url['director_name'],'director_url': dir_url}
            print(dir_url_dict)
            all_dir_urls.append(dir_url_dict)
        else:
            dir_url = None
            dir_url_dict = {'director_name': url['director_name'],'director_url': dir_url}
            print(dir_url_dict)
            print(f"Could not find 'href' attribute for {url}")
            all_dir_urls.append(dir_url_dict)
    else:
        dir_url = None
        dir_url_dict = {'director_name': url['director_name'],'director_url': dir_url}
        print(dir_url_dict)
        print(f"Could not find element with class 'ipc-metadata-list-summary-item__t' for {url}")
        all_dir_urls.append(dir_url_dict)
```

```python
urls_df = pd.DataFrame(all_dir_urls)

urls_df.to_csv('urls_df.csv', index=False)
```

```python
dir_birth_info = []
all_birth_info = []


for single_dir_url in all_dir_urls:
    try:
        search_url_for_requests = single_dir_url.get('director_url')
        response = requests.get(search_url_for_requests, headers=headers)
        soup = BeautifulSoup(response.text, 'html.parser')
        
        name_element = soup.find('span', class_="hero__primary-text")
        name_in_dir_page = name_element.text if name_element else None
    
        tag_born = soup.find(class_ = "ipc-metadata-list-item__label",string = re.compile('Born'))
        tag_next_container = tag_born.next.next
        tag_birth_origin =  tag_next_container.find(class_="ipc-metadata-list-item__list-content-item")
        birth_location = tag_birth_origin.text.split(", ")[-1]
        
        dir_birth_info = {'name_in_og_dataset': single_dir_url['director_name'], 'name_in_dir_page': name_in_dir_page, 'birth_location': birth_location}
        print(dir_birth_info)
        all_birth_info.append(dir_birth_info)
        
    except:
        birth_location = "Help! Couldn't find!"
        dir_birth_info = {'name_in_og_dataset': single_dir_url['director_name'], 'name_in_dir_page': name_in_dir_page, 'birth_location': birth_location}
        print(dir_birth_info)
        all_birth_info.append(dir_birth_info)
```
